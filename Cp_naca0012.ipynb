```python
# Зчитування координат (x; Cpi) з файлу Cp_6.0.txt та наступна побудова графіка
import math
import numpy as np
import matplotlib.pyplot as plt

L2='cp\Cp_6.0.txt' # назва файлу (його адреса);
f=6.0 #поворот на f°

# відкриваємо файл для зчитування даних
f2 = open(L2, mode='r')
X2,Y2=[],[]    
i=0
for row in f2:
    if i>5 and i<205:
        s=row.split(' ')
        if len(s[5])>1:
            s=[s[1], s[5]]
        elif len(s[4])>1:
            s=[s[1], s[4]]        
        s1=list(map(float, s))
        X2.append(s1[0])
        Y2.append(-s1[1])
    i=i+1
f2.close()

# будуємо з зчитаних даних профіль
plt.plot(X2, Y2, '-om',label='Cp, '+str(round(f,1))+'°', markersize=1)

plt.plot(0, 0, '-oy', markersize=3) #центр СК
plt.annotate('центр СК', xy=(0,0), fontsize=7)
plt.ylabel('-Cp')
plt.xlabel('x')
plt.title('NACA 0012')
plt.axis('equal')#однакові масштаби для x та y
plt.legend()
plt.ylim(-1.2, 2.8)
plt.xlim(-0.5, 1.5)
plt.show()

# 'b' blue
# 'g' green
# 'r' red
# 'c' cyan
# 'm' magenta
# 'y' yellow
# 'k' black
# 'w' white

```
