```python
import numpy as np
import matplotlib.pyplot as plt
#R=5
x0=-0.6
y0=2  #центр кола
R=(y0**2+(-x0+1)**2)**0.5 #радіус такого кола щоб воно проходило через точку(1; 0)
x=np.linspace(x0-R, x0+R, 50) #точки х кола(k)
vx=np.linspace(-2*R, 2*R, 50)
vy=np.linspace(-2*R, 2*R, 50) #для вісі координат
v0=[0]*50

# x=[-R, -0.9*R, -0.8*R, -0.7*R, -0.6*R, -0.5*R, -0.4*R, -0.3*R, -0.2*R, -0.1*R, -0.0*R, 0.1*R, 0.2*R, 0.3*R, 0.4*R, 0.5*R, 0.6*R, 0.7*R, 0.8*R, 0.9*R, R]
y1=[]
y2=[]
X1=[]
Y1=[]
X2=[]
Y2=[]

for i in range(len(x)):
    y1.append(((R**2-(x[i]-x0)**2)**0.5)+y0)
    y2.append((-(R**2-(x[i]-x0)**2)**0.5)+y0) #точки у кола(p)

    R1=x[i]**2-y1[i]**2+1
    M1=2*x[i]*y1[i]
    X1.append((R1*x[i]+M1*y1[i])/(x[i]**2+y1[i]**2))
    Y1.append((M1*x[i]-R1*y1[i])/(x[i]**2+y1[i]**2))
    
    R2=x[i]**2-y2[i]**2+1
    M2=2*x[i]*y2[i]
    X2.append((R2*x[i]+M2*y2[i])/(x[i]**2+y2[i]**2))
    Y2.append((M2*x[i]-R2*y2[i])/(x[i]**2+y2[i]**2))
print(len(y1))
print(len (x))
print(v0)
plt.axis('equal') #пропорційний графік
plt.plot(vx, v0, '-og')
plt.plot(v0, vy, '-og') #вісь іксів та вісь ігриків
plt.plot(x, y1, '-ob')
plt.plot(x, y2, '-oy')#будуємо графік
plt.plot(x0, y0, '-og')
plt.plot(1, 0, '-or')
plt.plot(X1, Y1, '-or')
plt.plot(X2, Y2, '-or')
plt.show() #показуємо графік
```
