```python
# Зчитування координат (x; y) з файлу NACA 0012.dat та наступна побудова профілю
import math
import numpy as np
import matplotlib.pyplot as plt

L1='NACA 0012.dat' # назва файлу (його адреса)
f=30 #поворот на f° відносно центру хорди

X1,Y1=[],[]
# відкриваємо файл для зчитування даних
f1 = open(L1, mode='r')
i=0
for row in f1:
    if i>0:
        s=row.split(' ')
        if len(s)==7:
            s=[s[1], s[6]]
        else:
            s=[s[1], s[5]]        
        s1=list(map(float, s))
        X1.append(s1[0])
        Y1.append(s1[1])
    i=i+1
f1.close()
# будуємо з зчитаних даних профіль
plt.plot(X1, Y1, '-og',label='профіль, 0°', markersize=1)

# поворот на f° та зміщення на 0.5 СК
f=f*np.pi/180 #f у радіанах
X11=np.array(X1)
Y11=np.array(Y1)
X1=(X11-0.5)*np.cos(f)+Y11*np.sin(f)+0.5
Y1=Y11*np.cos(f)-(X11-0.5)*np.sin(f)

# будуємо з зчитаних даних профіль повернений на f°
plt.plot(X1, Y1, '-oy',label='профіль, '+str(round(f*180/np.pi,1))+'°', markersize=1)

plt.plot(0, 0, '-or', markersize=3) #центр СК
plt.annotate('центр СК', xy=(0,0), fontsize=7)
plt.plot(0.5, 0, '-ob', markersize=3) #центр профілю
plt.annotate('центр профілю', xy=(0.5,0), fontsize=7)
plt.ylabel('y')
plt.xlabel('x')
plt.title('NACA 0012')
plt.axis('equal')#однакові масштаби для x та y
plt.legend()
plt.ylim(-0.4, 0.4)
plt.xlim(-0.5, 1.5)
plt.show()

# 'b' blue
# 'g' green
# 'r' red
# 'c' cyan
# 'm' magenta
# 'y' yellow
# 'k' black
# 'w' white
```
